```python
from torch.utils.data import Dataset, DataLoader
import torch
import pandas as pd
from skimage import io, transform
import os

class unbc(Dataset):

    def __init__(self, csv_file, train=True, transform=None):

        self.landmarks_frame = csv_file
        self.transform = transform
        self.train = train
        self.data = load_data(csv_file,train)


    def __len__(self):
        return len(self.landmarks_frame)

    def __getitem__(self, idx):
        if torch.is_tensor(idx):
            idx = idx.tolist()

        img_name = os.path.join(self.root_dir,
                                self.landmarks_frame.iloc[idx, 0])
        image = io.imread(img_name)
        landmarks = self.landmarks_frame.iloc[idx, 1:]
        landmarks = np.array([landmarks], dtype=float).reshape(-1, 2)
        sample = {'image': image, 'landmarks': landmarks}

        if self.transform:
            sample = self.transform(sample)

        return sample
    def load_data(data_dir, train):
      # Load CSV file
      label_frame = read_path(data_dir, train)
      data = []
      color=0
      for idx, path in enumerate(label_frame):
          img = np.expand_dims(cv2.imread(path, color), -1) if color == 0 else cv2.imread(path, color)
          data.append(img)
          if not idx % 1000:
              print("{} images processed".format(idx))
      data = np.array(data, dtype=np.float32)
      labels = np.array(read_labels(data_dir, train))
      return data, labels


    def read_path(file_path, train):
        # Specify the path to your CSV file
        csv_file_path = file_path
        train=train
        data = []

        with open(csv_file_path, 'r') as csv_file:
            csv_reader = csv.reader(csv_file)

            for row in csv_reader:
                indx = int(row[0])
                path = row[1]
                if train:
                    if 6 <= indx <= 25:
                        data.append(path)
                else:
                    if 1 <= indx <= 5:
                        data.append(path)

        # Join the values into strings
        #data_string = ','.join(data)

        #return data_string
        print(data[:1000])
        return data
    def read_labels (file_path, train):
        # Specify the path to your CSV file
        csv_file_path = file_path
        train=train
        labels = []

        with open(csv_file_path, 'r') as csv_file:
            csv_reader = csv.reader(csv_file)
            for row in csv_reader:
                indx = int(row[0])
                path = row[2]
                if train:
                    if 6 <= indx <= 25:
                        labels.append(path)
                else:
                    if 1 <= indx <= 5:
                        labels.append(path)

        # Join the values into a string
        #labels_string = ' '.join(labels)

        return labels

data_dir = '/content/drive/MyDrive/Thesis/Federated-Learning-PyTorch/data/unbc/test-img-lab.csv'
train_dataset = unbc(data_dir, train=True, transform=transform)
print (train_dataset)
```
